```python
import pandas as pd

# Load the datasets
users = pd.read_csv('data/users.csv')
fake_users = pd.read_csv('data/fusers.csv')

# Add labels
users['is_fake'] = 0  # Real users labeled as 0
fake_users['is_fake'] = 1  # Fake users labeled as 1

# Combine both datasets
data = pd.concat([users, fake_users], ignore_index=True)

# Display dataset shape and first few rows
print(data.shape)
data.head()
```

```python
print(data.isnull().sum())
```

```python
import pandas as pd

# Specify the datetime format for 'updated'
updated_datetime_format = "%Y-%m-%d %H:%M:%S"

# Convert 'created_at' with a flexible parser (handles timezone info automatically)
data['created_at'] = pd.to_datetime(data['created_at'], format=None, errors='coerce', utc=True)

# Convert 'updated' with the specified format (naive)
data['updated'] = pd.to_datetime(data['updated'], format=updated_datetime_format, errors='coerce')

# Localize 'updated' to UTC if it's naive
if data['updated'].dt.tz is None:
    data['updated'] = data['updated'].dt.tz_localize('UTC')

# Extract features from datetime columns
data['created_year'] = data['created_at'].dt.year
data['created_month'] = data['created_at'].dt.month
data['created_day'] = data['created_at'].dt.day
data['created_hour'] = data['created_at'].dt.hour

data['updated_year'] = data['updated'].dt.year
data['updated_month'] = data['updated'].dt.month
data['updated_day'] = data['updated'].dt.day

# Calculate the difference in days between created and updated
data['days_diff'] = (data['updated'] - data['created_at']).dt.days

print(data)
```

```python
data.head(4)
```

```python
data = data.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
data.head()
```

```python
# Keep only the necessary columns
features = [
    'friends_count','statuses_count','followers_count','favourites_count', 'listed_count', 
    'favourites_count','days_diff'
]
target = 'is_fake'

X = data[features]
y = data[target]
```

```python
data[features]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, f1_score

# Make predictions
y_pred = model.predict(X_test)

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")

# Classification Report
print("Classification Report:\n", classification_report(y_test, y_pred))

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", conf_matrix)

# F1 Score (macro-averaged)
f1 = f1_score(y_test, y_pred, average='macro')
print(f"F1 Score (Macro): {f1:.4f}")
```
